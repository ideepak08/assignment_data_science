2. What is the difference between string and variable?
Answer:
    Variable- Variable is an identifier (or some kind of means) to store a value in a memory location and to manipulate it if 
    needed.
    my_var = 10
    x = 20
    String- Anything you assign to a variable using quotes (single or double) will be considered as a string. Thus, a string is
    one of several types of a variable.
    my_var = "10"
    x = "Deepak"

8. Why is eggs a valid variable name while 100 is invalid?
Answer: 
    Python allows you to name variables to your liking, as long as the names follow these rules:
        a. Variable names may contain letters, digits (0-9) or the underscore character _.
        b. Variable names must begin with a letter from A-Z or the underscore _ character. Either lowercase or uppercase letters are acceptable.
        c. Variable names may not be a reserved word in Python.
    Here egg fit in all criteria but 100 voilates 2nd condition.
